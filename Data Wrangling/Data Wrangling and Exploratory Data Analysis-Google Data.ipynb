## 1. Handle missing values

```python
for i in df.columns:    
    nans=df[i].isnull().sum()
    if nans:
        print (i,nans,'\n') 
```

```python
df["Textuelle Bewertung"].describe(include='object')[0]-df.shape[0]
```

```python
# simply drop whole row with NaN in "Textuelle Bewertung" column(because target value)
df.dropna(subset=["Textuelle Bewertung"], axis=0, inplace=True)

# reset index, because we droped two rows
df.reset_index(drop=True, inplace=True)
```

```python
import textblob
from textblob_de import TextBlobDE
```

```python
comments=df["Textuelle Bewertung"].tolist()
comments[:2]
```

```python
print("Beispiel...")
for i in comments[:3]:
    print("polarity: ",TextBlobDE(i).sentiment[0])
    print("subjectivity: ",TextBlobDE(i).sentiment[1])
```

```python
polaritiy=[TextBlobDE(i).sentiment[0] for i in comments]
subjectivity=[TextBlobDE(i).sentiment[1] for i in comments]
df['Polaritiy'] = polaritiy
df['Subjectivity'] = subjectivity
df.head()
```

```python
from datetime import datetime
```

```python
this_year=datetime.now().year
this_year
```

```python
this_month=datetime.now().month
this_month
```

```python
for i in df["Datum der Bewertung"]:
    i=i.split()
    del i[0]
    
    if ((i[0]=='einem') | (i[0]=='einer')):
        i[0]=1
    
    i[0]=int(i[0])
    
    if ((i[1]=='Jahr') | (i[1]=='Jahren')):
        i[1] = this_year - i[0]
        
    if ((i[1]=='Monaten') | (i[1]=='Monat')):
        if i[0]>=this_month:
            i[1] = this_year - 1
        else:
            i[1] = this_year

    if ((i[1]=='Wochen') | (i[1]=='Woche') | 
        (i[1]=='Tagen') | (i[1]=='Tag') |
        (i[1]=='Stunden') | (i[1]=='Stunde') |
        (i[1]=='Minuten') | (i[1]=='Minute') | 
        (i[1]=='Sekunden') | (i[1]=='Sekunde')):
        i[1] = this_year
    
    
    if type(i[1]) != type(34):
        print(i[1])
    print(i[1])
```

```python
for i in range(df.shape[0]):
    d=df.xs(i)["Datum"]
    d=d.split(".")
    df.at[i, 'Datum']=[d[1],d[0],d[2]]
    df.at[i, 'Datum']='/'.join(df.xs(i)["Datum"])
    print(df.xs(i)["Datum"])
```

<h2 id="data_standardization">3. Data Standardization</h2>

'Gesamtzufriedenheit', 'Qualit√§t der Beratung', 'Mediz. Behandlung','Verwaltung und Abl√§ufe', 'Ausstattung und Gestaltung'

Diese sollen normalerweise Number sein.

```python
bewertunglist=['Gesamtzufriedenheit','Qualit√§t der Beratung','Mediz. Behandlung','Verwaltung und Abl√§ufe','Ausstattung und Gestaltung']
sternelist=['Gesamt','Qualit√§t','Behandlung','Verwaltung','Ausstattung']
```

```python
for b in range(len(bewertunglist)):    
    df.loc[df[bewertunglist[b]] =="sehr zufrieden", sternelist[b]] = 100
    df.loc[df[bewertunglist[b]] =="zufrieden", sternelist[b]] = 67
    df.loc[df[bewertunglist[b]] =="weniger zufrieden", sternelist[b]] = 33
    df.loc[df[bewertunglist[b]] =="unzufrieden", sternelist[b]] = 0
```
